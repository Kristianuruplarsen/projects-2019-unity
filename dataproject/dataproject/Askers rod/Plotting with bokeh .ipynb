```python
import datetime
import pandas as pd
import numpy as np


#imports for plot
from bokeh.io import output_notebook, show, push_notebook
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, Select

from bokeh.io import curdoc
# imports for interaction
from bokeh.layouts import widgetbox, row
from bokeh.models.widgets import Select

from ipywidgets import interact
```

```python
#This tels python to show the output in this notebook instead of in a html file
output_notebook()
```

```python
#data import, from the HDF5-format:
d = pd.read_hdf('data.h5', key='losses')

# for later reference we make lists of our options
company_list = list(d.columns.levels[1])
attribute_list = list(d.columns.levels[0])
```

```python
close = ColumnDataSource(data= {'x' : np.array(d.index), 'y' : np.array(d['Close']['GOOG'])})
```

# Plotting the closing price only

```python
p = figure(x_axis_type='datetime',title=f'Graph of clossing price', \
    tools="pan,box_zoom,reset,save", y_axis_label='Closing price', x_axis_label='Date')
p.title.text = 'Closing price of GOOGLE'

p.line(x='x', y='y', source=close, color = 'blue', legend='close')


def update_name(company):
    new_close = {'x' : np.array(d.index),'y' : np.array(d['Close'][company])}
    close.data = new_close
    p.title.text = 'Closing price of ' + str(company)
    push_notebook()
    
show(p,notebook_handle=True)


interact(update_name,company=company_list)
```
