```python
import datetime
import pandas as pd
import numpy as np


#imports for plot
from bokeh.io import output_notebook, show, push_notebook
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, Select

from bokeh.io import curdoc
# imports for interaction
from bokeh.layouts import widgetbox, row
from bokeh.models.widgets import Select

from ipywidgets import interact
```

```python
#This tels python to show the output in this notebook instead of in a html file
output_notebook()
```

```python
#data import, from the HDF5-format:
d = pd.read_hdf('data.h5', key='losses')

# for later reference we make lists of our options
company_list = list(d.columns.levels[0])
attribute_list = list(d.columns.levels[1])
```

```python
close = ColumnDataSource(data= {'x' : np.array(d.index), 'y' : np.array(d['GOOG']['Close'])})
```

# Plotting the closing price only

```python
p = figure(x_axis_type='datetime',title=f'Graph of clossing price', \
    tools="pan,box_zoom,reset,save", y_axis_label='Closing price', x_axis_label='Date')
p.title.text = 'Closing price of GOOGLE'

p.line(x='x', y='y', source=close, color = 'blue', legend='Closing price')


def update_name(company):
    new_close = {'x' : np.array(d.index),'y' : np.array(d[company]['Close'])}
    close.data = new_close
    p.title.text = 'Closing price of ' + str(company)
    push_notebook()
p.legend.location = "top_left"    
show(p,notebook_handle=True)


interact(update_name,company=company_list)
```

# Plotting running means also

```python
company = 'GOOG'
companysource = ColumnDataSource(d[company])
p = figure(x_axis_type='datetime',title=f'Graph with running means for {company}', \
    tools="pan,box_zoom,reset,save", y_axis_label='Closing price', x_axis_label='Date')

p.line(x='Date', y='Close', source=companysource, color = 'blue', legend= 'Closing price')
p.line(x='Date', y='rm_200', source=companysource, color='black', line_dash='4 4', legend = '200 days')
p.line(x='Date', y='rm_50', source=companysource, color='red', line_dash='4 4',legend = '50 days')
p.legend.location = "top_left"


def update_name(company):
    comp = str(company)
    p.title.text = 'Closing price and running mean of ' + str(company)
    companysource.data = ColumnDataSource(d[str(company)]).data
    push_notebook()
    
show(p,notebook_handle=True)


interact(update_name,company=company_list)
```
